```python
import matplotlib.pyplot as plt
import numpy

# Sample data for the trajectory (x and y coordinates)
x =  numpy.load("../traj.npy")
gt = numpy.load("../gt.npy")

# Create a 2D trajectory plot
plt.figure(figsize=(20, 20))  # Optional: Set the figure size
plt.plot(x[:, 0], x[:, 1], marker='o', linestyle='-', color='b', label='Trajectory')
plt.plot(gt[:, 0], gt[:, 1], marker='o', linestyle='-', color='b', label='Trajectory')
plt.title('2D Trajectory Plot')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.grid(True)
plt.legend()
plt.show()
```
